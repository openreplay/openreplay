```python
#TODO: Find out why when adding user_id clickhouse dont show result
X,y,d = get_training_database()
```

```python
import mlflow
tracking_uri = f"postgresql+psycopg2://{_pg_user}:{_pg_password}@{_pg_host}:{_pg_port}/mlruns"
mlflow.set_tracking_uri(tracking_uri)
os.environ["MLFLOW_TRACKING_URI"] = tracking_uri
```

```python
from datetime import date
```

```python
#TODO: Set artifact uri to custom place (local, s3, etc)
#mlflow.set_experiment('svm-recommendation')
artifact_uri = 's3://openreplay-ml/recommendations/'
mlflow.create_experiment('s3-recommendations', artifact_uri)
mlflow.set_experiment('s3-recommendations')
with mlflow.start_run(run_name=f'experiment-{date.today().strftime("%Y-%M-%d_%H:%m")}'):
    reg_model_name = "SVM"
    best_meta = {'score': 0, 'model': None, 'name': 'NoName'}
    for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
        with mlflow.start_run(run_name=f'sub_run_with_{kernel}', nested=True):
            print("--")
            model = SVM_recommendation(kernel=kernel, test=True)
            mlflow.sklearn.autolog()
            model.fit(X, y)
            mlflow.log_param("kernel", kernel)
            mlflow.log_metric("score", model.score)
            if model.score > best_meta['score']:
                best_meta['score'] = model.score
                best_meta['model'] = model
                best_meta['name'] = kernel
    mlflow.log_metric("score", best_meta['score'])
    mlflow.log_param("name", best_meta['name'])
    mlflow.sklearn.log_model(best_meta['model'], "sk_learn",
                             serialization_format="cloudpickle",
                             registered_model_name=reg_model_name,
                             #artifact_path='s3://openreplay-quickwit/ml/'
                            )
```

```python
#del model
model = mlflow.pyfunc.load_model(f'models:/{reg_model_name}/1')
```

```python
pred = model.predict(X)
```

```python
acc = 0
for i in range(len(pred)):
    if pred[i] == y[i]:
        acc += 1
print(f'Accuracy: {acc/len(pred)}')
```
